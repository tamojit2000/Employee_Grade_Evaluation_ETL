```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window

Source_Data='/Volumes/workspace/data/employee_etl/Source_Data'
Temp='/Volumes/workspace/data/employee_etl/tmp'
Output='/Volumes/workspace/data/employee_etl/Output'
```

```python
df_attd=spark.read.csv(Source_Data+'/employee_attendance.csv',header=True,inferSchema=True).orderBy(col('Emp_Id'),col('Entry'))
df_attd.display()
```

```python
df_attd=df_attd.withColumn('Hours_Per_Day', round((unix_timestamp(col('Exit')) - unix_timestamp(col('Entry'))) / 3600,2)).filter(col('Hours_per_day') >= 8)
df_attd.display()
```

```python
df_pto=spark.read.csv(Source_Data+'/employee_pto.csv',header=True,inferSchema=True)
df_pto.display()
```

```python
df_pto=df_pto.withColumn("Entry", col("Date")) \
    .withColumn("Exit", col("Date")) \
    .withColumn("Hours_Per_Day", lit(8)) \
    .drop(col("Date"))
    
df_pto.display()
```

```python
df_attd=df_attd.unionByName(df_pto)
df_attd=df_attd.orderBy(col("Emp_Id"),col("Entry"))
df_attd.display()
```

```python
del df_pto
```

```python
df_attd=df_attd.withColumn('Week_Name',weekofyear(col('Entry'))).orderBy(col('Emp_Id'),col("Entry"))
df_attd.display()
```

```python
df_attd=df_attd.withColumn('Week_Name',weekofyear(col('Entry')))\
    .groupBy("Emp_Id","Week_Name")\
    .agg(count("*").alias("Total_Attendance_Per_Week"))\
        .orderBy(col('Emp_Id'),col('Week_Name'))

df_attd.display()
```

```python
df_attd=df_attd.filter(col("Total_Attendance_Per_Week")>=3)
df_attd.display()
```

```python
df_attd=df_attd.groupBy(col("Emp_Id")).agg(count('*').alias('cnt')).filter(col('cnt')>=4).select("Emp_Id")
df_attd=df_attd.withColumn('RTO',lit('Complaint'))
df_attd.display()
```

```python
df_attd.write.mode('overwrite').parquet(Temp+'/RTO_LRF.parquet')
```
