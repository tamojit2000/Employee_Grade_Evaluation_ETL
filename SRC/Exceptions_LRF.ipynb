```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window

Source_Data='/Volumes/workspace/data/employee_etl/Source_Data'
Temp='/Volumes/workspace/data/employee_etl/tmp'
Output='/Volumes/workspace/data/employee_etl/Output'
```

```python
df_sup=spark.read.parquet(Temp+'/Suspression_LRF.parquet',header=True,inferSchema=True)
df_sup.display()
```

```python
df_excep=spark.read.parquet(Temp+'/Negative_LRF.parquet',header=True,inferSchema=True)
df_excep.display()
```
